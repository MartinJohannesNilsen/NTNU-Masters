```python
import numpy as np
import pandas as pd
from pathlib import Path
import sys
import os

not_preprocessed = Path(os.path.abspath("")).parents[1] / "features" / "liwc" / "not_preprocessed" / "original"
```

```python
train = pd.read_csv(not_preprocessed / "csv" / "2022" / "train_sliced_stair_twitter.csv")
test = pd.read_csv(not_preprocessed / "csv" / "2022" / "test_sliced_stair_twitter.csv")
val = pd.read_csv(not_preprocessed / "csv" / "2022" / "shooter_hold_out_test.csv")

all_df = pd.concat([train, test, val], axis=0)
shooters = all_df[all_df["label"] == 1].iloc[:, 4:]
non_shooters = all_df[all_df["label"] == 0].iloc[:, 4:]
```

```python
print(len(shooters.columns))
print(shooters.columns)
```

```python
avg_cols = shooters.mean(axis=0)
#print(avg_cols)
#avg_cols = avg_cols.sort_values(ascending=False)[:15]
#print(avg_cols)
```

```python
avg_cols_non = non_shooters.mean(axis=0)
#print(avg_cols)
#avg_cols_non = avg_cols_non.sort_values(ascending=False)[:15]
#print(avg_cols)
```

```python
new_cols = [(avg_cols[i] - avg_cols_non[i]) for i in range(len(avg_cols))]
print(new_cols)
```

```python
new_df = pd.DataFrame([new_cols], columns=shooters.columns)
```

```python
df_dict = new_df.to_dict("index")
df_dict = df_dict[0]
print(df_dict)
```

```python
sorted_keys = sorted(df_dict, key=df_dict.get, reverse=True)
for r in sorted_keys[:10]:
    print(r, df_dict[r])
```
