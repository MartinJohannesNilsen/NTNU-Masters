## Preprocessing

This file is intended for the generation of preprocessing pipelines. There will be two pipelines, one standard (advanced) and one simplistic. The reason for this is that we want a standardized preprocessing procedure where all the necessary steps are taken. In addition, there is a high chance of us using the BERT architectue, and here there is only a few of the steps we would like to perform, as the models tend to apply their own preprocessing/tokenizer as well.

### Imports

```python
from pathlib import Path
import os
import pandas as pd
from typing import List
import nltk
nltk.download('popular')
```

### Read data

```python
from csv import QUOTE_NONE

base_path = Path(os.path.abspath("")).parent / "dataset" / "outputs"
datasets = {
    "school_shooters": base_path / "school_shooters.csv",
    "manifestos": base_path / "manifestos.csv",
    "stair_twitter_archive": base_path / "stair_twitter_archive.csv",
    "twitter": base_path / "twitter.csv",
}

df = pd.read_csv(datasets["school_shooters"], encoding="utf-8", delimiter="‎", engine="python", quoting=QUOTE_NONE)
df
```

### Extract text

```python
text_example = df["text"].iloc[550]
text_example
```

### Preprocessing

```python
from nltk.tokenize import word_tokenize
def tokenization(input: str):
    return word_tokenize(input)

from nltk.corpus import stopwords
def stopword_elimination(tokenized_input: List):
    return [w for w in tokenized_input if not w.lower() in set(stopwords.words('english'))]

import string
punctuation = [e for e in string.punctuation] + ["....", "—", "’"]
def punctuation_elimination(tokenized_input: List):
    return [t for t in tokenized_input if not t in punctuation]

# standard_preprocessing_steps = [tokenization]
standard_preprocessing_steps = [tokenization, stopword_elimination, punctuation_elimination]
# res = "... !! ?! _ - ? ! = : ~ @ demo test and"
res = text_example
for func in standard_preprocessing_steps:
    res = func(res)

res
```
