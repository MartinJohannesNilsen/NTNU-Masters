```python
from pathlib import Path
import os
import datetime as dt
import pandas as pd
```

### Set path

```python
doc_path = Path(os.path.abspath("")).parents[1] / "schoolshootersinfo" / "William_Atchison" / "data.csv"
assert os.path.isfile(doc_path), "Wrong path"
```

### Read data and append date columns

NB! Må fikse datetime der month eller day ikke er definert. Nå mistenker jeg at de blir 01.01.xxxx, som er feil. Kan vi få disse til å bli ingenting? Splitte på "-" og sette år, måned, dag hardt kanskje? Drite i datetime objekt en mulighet

```python
df = pd.read_csv(doc_path, delimiter="‎ ", engine='python')

# Change date column to datetime
df = df.rename(columns={"date": "datetime_utc"})
df["datetime_utc"] = pd.to_datetime(df["datetime_utc"])
# df.drop(["date"], axis=1)

# Append columns for year, month and day
df['year'] = df['datetime_utc'].dt.year.convert_dtypes()
df['month'] = df['datetime_utc'].dt.month.convert_dtypes()
df['day'] = df['datetime_utc'].dt.day.convert_dtypes()

df
```
