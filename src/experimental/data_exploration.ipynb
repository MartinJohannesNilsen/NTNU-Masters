### Read data and append date columns

NB! Må fikse datetime der month eller day ikke er definert. Nå mistenker jeg at de blir 01.01.xxxx, som er feil. Kan vi få disse til å bli ingenting? Splitte på "-" og sette år, måned, dag hardt kanskje? Drite i datetime objekt en mulighet

```python
# Read pdf
df = pd.read_csv(doc_path, delimiter="‎ ", engine='python')

# Append year, month and day columns. 
# This method insert None for undefined values
date_dict = { "year": [], "month": [], "day": [] }
for date in df["date"].convert_dtypes():
    year, month, day = [None, None, None] if type(date) == pd._libs.missing.NAType else (date.split("-") + 3 * [None])[:3]
    date_dict["year"].append(year)
    date_dict["month"].append(month)
    date_dict["day"].append(day)
for key in date_dict.keys():
    df[key] = date_dict[key]
df
```
