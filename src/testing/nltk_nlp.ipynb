```python
# Imports
import pandas as pd
import numpy as np
import nltk
import csv
import os

from nltk.tokenize import word_tokenize
from pathlib import Path
from nltk.probability import FreqDist
```

```python
# Setup
#nltk.download()
```

```python

from csv import QUOTE_NONE

paths = dict()
for csv in (Path(os.path.abspath("")).parents[1] / "data" / "school_shooters").rglob("data.csv"):
    paths[csv.parents[0].name] = csv

dfs = dict()
for key, value in paths.items():
    dfs[key] = pd.read_csv(value, encoding="utf-8", delimiter="‎", engine="python", quoting=QUOTE_NONE)

dfs
```

```python
df = dfs["Adam_Lanza"]
df
```

```python
all_posts_tokenized = []

for post in df["text"]:
    all_posts_tokenized.append(word_tokenize(post))

#all_posts_tokenized.flatten()
all_posts_tokenized = sum(all_posts_tokenized, [])#np.asarray(all_posts_tokenized, dtype="object").flatten()
word_frequencies = FreqDist(all_posts_tokenized)
```
