```python
import os
from pathlib import Path
import pandas as pd
from csv import QUOTE_NONE
import numpy as np
```

```python
data_dir = Path(os.path.abspath("")).parents[1] / "dataset_creation" / "data" / "train_test"
emb_dir = Path(os.path.abspath("")).parents[1] / "experiments" / "features" / "embeddings"
```

```python
# Files for comparison
data = {
    "hold_out": data_dir / "shooter_hold_out_test.csv",
    "train": data_dir / "train_sliced_stair_twitter.csv",
    "test": data_dir / "test_sliced_stair_twitter.csv"
}

embeddings = {
    "glove": {
        "hold_out": emb_dir / "hold_out_test_sliced_stair_twitter_glove.h5",
        "train": emb_dir / "train_sliced_stair_twitter_glove.h5",
        "test": emb_dir / "test_sliced_stair_twitter_glove.h5"
    },
    "fasttext": {
        "hold_out": emb_dir / "hold_out_test_sliced_stair_twitter_fasttext.h5",
        "train": emb_dir / "train_sliced_stair_twitter_bert_fasttext.h5",
        "test": emb_dir / "test_sliced_stair_twitter_bert_fasttext.h5"
    },
    "bert": {
        "hold_out": emb_dir / "hold_out_test_sliced_stair_twitter_bert.h5",
        "train": emb_dir / "train_sliced_stair_twitter_bert.h5",
        "test": emb_dir / "test_sliced_stair_twitter_bert.h5"
    },
}

lengths = {}
```

### Original files

```python

hold_out_df = pd.read_csv(data["hold_out"], encoding="utf-8", delimiter="‎", engine="python", quoting=QUOTE_NONE)
train_df = pd.read_csv(data["train"], encoding="utf-8", delimiter="‎", engine="python", quoting=QUOTE_NONE)
test_df = pd.read_csv(data["test"], encoding="utf-8", delimiter="‎", engine="python", quoting=QUOTE_NONE)

lengths["hold_out_original"] = len(hold_out_df.index)
lengths["train_original"] = len(train_df.index)
lengths["test_original"] = len(test_df.index)

print("hold out", lengths["hold_out_original"])
print("train", lengths["train_original"])
print("test", lengths["test_original"])
```
