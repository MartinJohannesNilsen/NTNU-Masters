```python
import numpy as np
import pandas as pd
from pathlib import Path
import sys
import os

not_preprocessed = Path(os.path.abspath("")).parents[1] / "features" / "liwc" / "original"
```

```python
train = pd.read_csv(not_preprocessed / "csv" / "2022" / "train_sliced_stair_twitter.csv")
test = pd.read_csv(not_preprocessed / "csv" / "2022" / "test_sliced_stair_twitter.csv")
val = pd.read_csv(not_preprocessed / "csv" / "2022" / "shooter_hold_out_test.csv")

all_df = pd.concat([train, test, val], axis=0)
shooters = all_df[all_df["label"] == 1]
non_shooters = all_df[all_df["label"] == 0]
```

```python
print(len(shooters.columns))
print(shooters.columns)
```

```python
avg_cols = shooters.mean(axis=0)
#print(avg_cols)
avg_cols = avg_cols.sort_values(ascending=False)[:15]
print(avg_cols)
```

```python
avg_cols_non = non_shooters.mean(axis=0)
#print(avg_cols)
avg_cols = avg_cols_non.sort_values(ascending=False)[:15]
print(avg_cols)
```
